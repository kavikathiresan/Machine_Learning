```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('T:\\python\\class\\50_Startups.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
len(df)
```

# convert categorical to numberical 

```python
df['State'].unique()
```

```python
df['State'].value_counts()
```

```python
# gives dummies categorical to numberic
```

```python
a = pd.get_dummies(df['State'],drop_first=True)
a
```

```python
# concat them a to df
```

```python
df=pd.concat([df,a],axis=1)
df
```

```python
# drop the state column
```

```python
df=df.drop(['State'],axis=1)
```

```python
df.head()
```

```python
# split the independent variable and dependent variable
```

```python
x = df.drop(['Profit'],axis=1)
x
```

```python
y = df['Profit']
y
```

```python
x.ndim
```

```python
x.shape
```

```python
# Training the data to train with Algorithm
```

```python
import sklearn
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
len(x_train),len(y_train)
```

```python
# Testing the data to test with Algorithm
```

```python
len(x_test) ,len(y_test)
```

```python
# data given to model
```

```python
import sklearn
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
reg.fit(x_train,y_train)
```

```python
# m value
```

```python
reg.coef_
```

```python
# c value
```

```python
reg.intercept_
```

```python
# simple linear regression
#y= mx+c
# y = 8.05630064e-01 x 8.05630064e-01 x1 + 2.98554429e-02 x2 + 9.38793006e+02 x3 + 6.98775997e+00 x4 + 54028.039593645844
```

```python
# lets trained model make predictions
```

```python
y_train_predict = reg.predict(x_train)
```

```python
y_train_predict 
```

```python
# finding the training loss
```

```python
from sklearn.metrics import r2_score
```

```python
print(f'the training model accurancy:{r2_score(y_train,y_train_predict)}')
```

```python
print(f'the loss training model :{1-r2_score(y_train,y_train_predict)}')
```
