```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
import warnings 
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv('T:/python/class/class_29/Mall_Customers.csv')
```

```python
df.head()
```

```python
X=df.iloc[:,3:5]
```

```python
from sklearn.cluster import KMeans
```

```python
K=range(1,10)
wcss=[]
for i in K:
    kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
    
```

```python
plt.plot(K,wcss,'bx--')
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# cluster = 5
kmeans=KMeans(n_clusters=5,init='k-means++',random_state=42)
ymeans=kmeans.fit_predict(X)
```

```python
centroid=kmeans.cluster_centers_
centroid
```

```python
plt.scatter(X.iloc[ymeans==0,0],X.iloc[ymeans==0,1],color='r')
plt.scatter(X.iloc[ymeans==1,0],X.iloc[ymeans==1,1],color='g')
plt.scatter(X.iloc[ymeans==2,0],X.iloc[ymeans==2,1],color='cyan')
plt.scatter(X.iloc[ymeans==3,0],X.iloc[ymeans==3,1],color='b')
plt.scatter(X.iloc[ymeans==4,0],X.iloc[ymeans==4,1],color='m')
plt.show()
```

```python
plt.scatter(X.iloc[ymeans == 0, 0], X.iloc[ymeans == 0, 1], s = 100, c = 'red', label = 'Careless Customers')
plt.scatter(X.iloc[ymeans == 1, 0], X.iloc[ymeans == 1, 1], s = 100, c = 'blue', label = 'Standard Customers')
plt.scatter(X.iloc[ymeans == 2, 0], X.iloc[ymeans == 2, 1], s = 100, c = 'green', label = 'Target Customers')
plt.scatter(X.iloc[ymeans == 3, 0], X.iloc[ymeans == 3, 1], s = 100, c = 'cyan', label = 'Sensible Customers')
plt.scatter(X.iloc[ymeans == 4, 0], X.iloc[ymeans == 4, 1], s = 100, c = 'magenta', label = 'Careful Customers')
plt.title('Clusters of Customers using K-Means Clustering')
plt.xlabel('Annual Income (K$)')
plt.ylabel('Spend Score (1-100)')
plt.legend()
plt.show()
```
