# OneHotEncoder using frameworks -classification and regression

```python
dfs = pd.read_csv('T:/python/class/class_23/customer.csv')
```

```python
dfs.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
sc=OneHotEncoder(drop='first')
res=sc.fit_transform(dfs['gender'].values.reshape(-1,1)).toarray().astype(int)# independent column and we are drop first value and it is a 1d array so convert to 2d array
res
```

```python
dfs['Gender_resplaced']=res
```

```python
dfs.head(10)
```

# OrdinalEncoder-classification and Regression

```python
from sklearn.preprocessing import OrdinalEncoder
oe=OrdinalEncoder()
result=oe.fit_transform(dfs[['review','education']])
result
```

```python
temp =pd.DataFrame(result,columns=['reviews','educations'])
temp
```

```python
dfs=pd.concat([dfs,temp],axis=1)
```

```python
dfs.head()
```

# Label Encoder -Classification (dependent variable only)

```python
from sklearn.preprocessing import LabelEncoder
l=LabelEncoder()
results=l.fit_transform(dfs['purchased'])
results
```

```python
dfs['Target']=results
```

```python
dfs.head()
```

```python
dfs=dfs.drop(['gender','review','education','purchased'],axis=1)
```

```python
dfs.head()
```

# Mean and Frequency Encoder-Regression datasets

```python
d=pd.read_csv('T:/python/class/class_23/house_price.csv')
```

```python
d.head()
```

```python
d.columns
```

```python
# label colum for mean encoder Technique
d['Neighborhood'].unique()
```

```python
d['Neighborhood'].value_counts()
```

```python
# finding mean value and groupby for neighrhood with target value
value=d.groupby(['Neighborhood'])['SalePrice'].mean()
value
```

```python
# to assign value we use dictionary
value=d.groupby(['Neighborhood'])['SalePrice'].mean().to_dict()
value
```

```python
d['Neighborhood']=d['Neighborhood'].map(value)
```
